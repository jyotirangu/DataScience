```python
# from a pandas dataframe 

batsman = pd.read_csv('1sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# plotting multiple plots            

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# labels title    

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])  

plt.title('Rohit Sharma Vs Virat Kohli Career Comparsion')
plt.xlabel('Season')
plt.ylabel('Runs Scored')        
```

```python
# colors(hex) and line(width and style) and marker(size) 

plt.plot(batsman['index'], batsman['V Kohli'], color='red')
plt.plot(batsman['index'], batsman['RG Sharma'],color='blue')  

plt.title('Rohit Sharma Vs Virat Kohli Career Comparsion')
plt.xlabel('Season')
plt.ylabel('Runs Scored') 
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='dashed')
plt.plot(batsman['index'], batsman['RG Sharma'],color='blue', linestyle='dashed')  

plt.title('Rohit Sharma Vs Virat Kohli Career Comparsion')
plt.xlabel('Season')
plt.ylabel('Runs Scored') 
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='dotted')
plt.plot(batsman['index'], batsman['RG Sharma'],color='blue', linestyle='dotted')  

plt.title('Rohit Sharma Vs Virat Kohli Career Comparsion')
plt.xlabel('Season')
plt.ylabel('Runs Scored') 
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='dashdot')
plt.plot(batsman['index'], batsman['RG Sharma'],color='blue', linestyle='dashdot')  

plt.title('Rohit Sharma Vs Virat Kohli Career Comparsion')
plt.xlabel('Season')
plt.ylabel('Runs Scored') 
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='solid',linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'],color='blue', linestyle='dotted',linewidth=2)  

plt.title('Rohit Sharma Vs Virat Kohli Career Comparsion')
plt.xlabel('Season')
plt.ylabel('Runs Scored') 
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'], batsman['RG Sharma'],color='blue', linestyle='dotted',linewidth=2,marker='D')  

plt.title('Rohit Sharma Vs Virat Kohli Career Comparsion')
plt.xlabel('Season')
plt.ylabel('Runs Scored') 
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='solid',linewidth=3,marker='D',markersize=10)
plt.plot(batsman['index'], batsman['RG Sharma'],color='blue', linestyle='dotted',linewidth=2,marker='D')  

plt.title('Rohit Sharma Vs Virat Kohli Career Comparsion')
plt.xlabel('Season')
plt.ylabel('Runs Scored') 
```

```python
# legend -> location 

plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='solid',linewidth=3,marker='D',markersize=10, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'],color='blue', linestyle='dotted',linewidth=2,marker='D', label='Rohit')  

plt.title('Rohit Sharma Vs Virat Kohli Career Comparsion')
plt.xlabel('Season')
plt.ylabel('Runs Scored') 

plt.legend()
```

```python
# in above graph the index is situated on top left side, thats becuase in plt.legend() there is the default value set as plt.legend(loc='best') which means the index will appeare whereever the free space available
```

```python
# limiting axes

price = [48000,39900,14999,45932,45493,29384,2394587]
year =[2019,2020,2021,2022,2023,2024,2025]

plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2019,2020)
```

```python
# grid 

plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'], batsman['RG Sharma'],color='blue', linestyle='dotted',linewidth=2,marker='D')  

plt.title('Rohit Sharma Vs Virat Kohli Career Comparsion')
plt.xlabel('Season')
plt.ylabel('Runs Scored') 

plt.grid()
```

```python
# show 

plt.plot(batsman['index'], batsman['V Kohli'], color='red', linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'], batsman['RG Sharma'],color='blue', linestyle='dotted',linewidth=2,marker='D')  

plt.title('Rohit Sharma Vs Virat Kohli Career Comparsion')
plt.xlabel('Season')
plt.ylabel('Runs Scored') 

plt.grid()

plt.show()
```

# Scatter Plots

image.png

- Bivariate Analysis

- Use case - Finding correlation

```python
# plt.scatter simple function 

x = np.linspace(-10,10,50)

y = 10*x + 3 + np.random.randint(0,300,50)
```

```python
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data  

df = pd.read_csv('1batter.csv')   
df      
```

```python
plt.scatter(df['avg'], df['strike_rate'],color='red')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
# marker 

plt.scatter(df['avg'], df['strike_rate'],color='red',marker='*')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('SR')
```

```python
# size 
tips = sns.load_dataset('tips')
tips
```

```python
# slower

plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

```python
# scatterplot using plt.plot 

# Faster             
plt.plot(tips['total_bill'],tips['tip'])
```

### Bar chart

image.png

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

```python
# simple bar chart   

children = [10,20,30,10,20]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,children,color='black')
```

```python
# horizontal bar chart   

plt.barh(colors,children,color='black')
```

```python
# color and label 

df = pd.read_csv('1batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'])
```

```python
plt.bar(df['batsman'],df['2015'])
plt.bar(df['batsman'],df['2016'])
plt.bar(df['batsman'],df['2017'])
```

```python
plt.bar(df['batsman'],df['2015'],width=0.5)
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'], width = 0.2)
plt.bar(np.arange(df.shape[0]),df['2016'], width = 0.2)
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'], width = 0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'])
plt.show()
```

```python
children = [10,20,30,10,20]
colors = ['redredredred','blueblueblueblue','greengreengreen','yellowyellowyellow','pinkpinkpinkpink']

plt.bar(colors,children,color='black')
```

```python
# in above graph you can see on the x axis the readability reduces to overcome this we will use xticks

children = [10,20,30,10,20]
colors = ['redredredred','blueblueblueblue','greengreengreen','yellowyellowyellow','pinkpinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical')
```

```python
# stack bar chart 
df
```

```python
plt.bar(df['batsman'],df['2017'], label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'], label='2016')
plt.bar(df['batsman'],df['2015'],bottom=(df['2016'] + df['2017']), label='2015')

plt.legend()
plt.show()
```

### Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
data = [23,45,67,58,7,9]
marks=['english','hindi','marathi','science','maths','history']
plt.pie(data,labels=marks)
plt.show()
```

```python
df = pd.read_csv('1gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
# percentage and color   

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors=['red','green','white','blue','orange'])
plt.show()      
```

```python
# explode shadow

plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.2,0,0,0,0,0],shadow=True)
plt.show()
```
