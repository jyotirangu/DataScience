```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=1)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

```python
# cmap and alpha 
```

##### cmap (Colormap):

A colormap is a mapping of data values to colors. It defines the range of colors used to represent different magnitudes or categories within your data. For example, a "viridis" colormap might map low values to dark purples and high values to bright yellows, while a "grayscale" colormap maps values to shades of gray. cmap is used to visualize the values of your data.

##### alpha (Transparency):

Alpha controls the transparency or opacity of plotted elements. It is a value between 0 (fully transparent) and 1 (fully opaque). Setting alpha to a value less than 1 allows underlying elements or the background to show through. alpha is used to control the visibility of your plotted elements, especially useful for overlapping data or reducing visual clutter.

```python
# resizing the graph 

plt.figure(figsize=(10,5))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=1)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

### Annotations

```python
batters = pd.read_csv('2batter.csv')
batters.head(5)
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
# in the below scatter graph we can not identify the specific strike rate for a particular average. So we need to add a legend to the graph to identify the batsman 
```

```python
plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'],sample_df['strike_rate'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

### Horizontal and Vertical lines

```python
plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i], fontdict={'size': 15, 'color':'black'})
```

### Subplots

```python
# A different way to plot graphs         

batters.head()
```

```python
# 1st way      

plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Heading')
plt.xlabel('Average')
plt.ylabel('Strike Rate')

plt.show()
```

```python
plt.subplots(figsize=(15,6))
```

```python
# in the above output
# (<Figure size 1500x600 with 1 Axes>, <Axes: >)
# <Figure size 1500x600 with 1 Axes> defines figure info and <Axes: > defines 
# axes info 
# The figure info is a tuple containing the figure size and the number of axes in the figure.
# The axes info is a tuple containing the axes info. The axes info is a tuple containing the axes info.
# The variable fig and ax are the figure and axes objects, respectively.
```

```python
# 2nd way   

fig, ax = plt.subplots(figsize=(15,6))

ax.scatter(batters['avg'], batters['strike_rate'])
ax.set_title('Heading')
ax.set_xlabel('Average')
ax.set_ylabel('Strike Rate')

fig.show()

# if both the ways generate same results, then what is the need for the two separate methods, that is because by using the second method we can generate two graphs simultaneously 
```

```python
plt.subplots(nrows=2,ncols=1)
```

```python
fig, ax = plt.subplots(nrows=2,ncols=1,sharex=True,figsize=(15,6))

ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
plt.subplots(nrows=2,ncols=2,sharex=True,figsize=(15,6))
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(18,10))

ax[0,0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0,1].scatter(batters['avg'],batters['runs'])
ax[1,0].hist(batters['avg'])
ax[1,1].hist(batters['runs'])
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2,2,1)
ax1.scatter(batters['avg'],batters['strike_rate'],color='red')

ax2 = fig.add_subplot(2,2,2)
ax2.scatter(batters['avg'],batters['runs'])

ax3 = fig.add_subplot(2,2,3)
ax3.hist(batters['runs'])

ax4 = fig.add_subplot(2,2,4)
ax4.hist(batters['avg'])
```

### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
x
```

```python
xx, yy = np.meshgrid(x, y)
```

```python
xx.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.log(xx) + np.log(yy)

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```
