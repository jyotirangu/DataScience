### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

### A second way to plot Facet plots -> FacetGrid

```python
# figure level -> relplot -> displot -> catplot -> lmplot

sns.catplot(data=tips, y='total_bill', x='sex', kind='violin',col='day',row='time')
```

we will plot the same above graph using facet grid

```python
sns.FacetGrid(data=tips, col='day', row='time')
```

```python
# Create the facet grid (define columns and rows for faceting)
g = sns.FacetGrid(data=tips, col='day', row='time')

# Then map a plot type to it, e.g., scatterplot or barplot
g.map(sns.violinplot, 'sex', 'total_bill')
```

Facet grid allows you to do lot of customization and it is very flexible. You can add or remove facets as per your requirement.

```python
# add legend to the facet grid

g = sns.FacetGrid(data=tips, col='day', row='time',hue='smoker')
g.map(sns.violinplot, 'sex', 'total_bill')
g.add_legend()
```

### Plotting Pairwise Relationship (PairGrid Vs Pairplot)

```python
iris
```

as you can see in the iris data there are 4 numerical columns and one categorical data, pairplot will detect the numerical columns and pair by pair plot scatter plot between them

```python
sns.pairplot(data=iris)
```

```python
# hue 

sns.pairplot(data=iris,hue='species')
```

internally pariplot use pairgrid to plot the data

```python
# pair grid 
g = sns.PairGrid(data=iris)

# g.map 
g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
```

You can draw diagonal data and off-diagonal data separately

```python
# map_diag -> map_offdiag

g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.scatterplot)
```

The advantage of the PairGrid is you can use different type of graph for dia and off-dia

```python
# map_diag -> map_upper -> map_lower

g = sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.histplot)
```

You can even restrict the columns on which you want to plot graph using pairgrid via. vars method

```python
# vars

g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width', 'petal_width'])
g.map_diag(sns.kdeplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.histplot)
```
